```python
from konlpy.tag import Kkma, Komoran, Okt, Hannanum
from konlpy.utils import pprint #보기 좋게 출력할 때 사용
han = Hannanum()
okt = Okt()
kkma = Kkma(8092)
komoran = Komoran()
```

```python
text = "우리가 아쉬운 만큼 본인도 아쉬웠겠지. 다른 데 돈 쓰는 건 별로 안 아까운데 우산 사는 건 아깝더라."
```

```python
# 형태소 출력
print('hannanum:\n', han.morphs(text), '\n')
print('Okt:\n', okt.morphs(text), '\n')
print('Kkma:\n', kkma.morphs(text), '\n')
print('Komoran:\n', komoran.morphs(text), '\n')
```

```python
# (형태소, tag) 출력
print('hannanum:\n', han.pos(text))      # Hannanum을 사용한 형태소 분석
print('Okt:\n', okt.pos(text))      # Okt을 사용한 형태소 분석
print('Kkma:\n', kkma.pos(text))     # Kkma을 사용한 형태소 분석
print('Komoran:\n', komoran.pos(text))  # Komoran을 사용한 형태소 분석
```

```python
# 형태소 중 명사만 출력
print('hannanum:\n', han.nouns(text))      # Hannanum을 사용한 형태소 분석
print('Okt:\n', okt.nouns(text))      # Okt을 사용한 형태소 분석
print('Kkma:\n', kkma.nouns(text))     # Kkma을 사용한 형태소 분석
print('Komoran:\n', komoran.nouns(text))  # Komoran을 사용한 형태소 분석
```

```python
# txt 파일 로드해서 형태소 분석 후 txt 파일로 저장(Kkma)
from konlpy.tag import Kkma

file_path = './output/NZRW2400000004.txt'
file_name = 'NZRW2400000004'
with open(file_path, 'r', encoding='utf-8') as file:
    textfile = file.read()

kkma = Kkma(max_heap_size=8092)
tagged = kkma.pos(textfile)
result = '\n'.join([f"{word}_{pos}" for word, pos in tagged])


with open(f'./tagged/{file_name}_tagged.txt', 'w', encoding='utf-8') as result_file:
    result_file.write(result)

print("형태소 분석 결과가 'POS_tagged_kkma.txt' 파일로 저장되었습니다.")
```

```python
import os
import zipfile
from konlpy.tag import Kkma

# 1. 경로 설정
zip_path = '/content/drive/MyDrive/실습_txt변환/textbook_sample.zip'
output_dir = '/content/drive/MyDrive/실습_txt변환/output_tagged/'
final_zip_name = '/content/drive/MyDrive/실습_txt변환/textbook_sample_tagged_kkma.zip'

# 결과를 저장할 임시 폴더 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

kkma = Kkma()

# 2. ZIP 파일 읽기 및 분석
with zipfile.ZipFile(zip_path, 'r') as z:
    # zip 안의 파일 목록 중 .txt 파일만 추출
    txt_files = [f for f in z.namelist() if f.endswith('.txt')]

    for filename in txt_files:
        print(f"분석 중: {filename}...")

        # 파일 내용 읽기
        with z.open(filename) as f:
            textfile = f.read().decode('utf-8')

        try:
            # 형태소 분석 (Kkma)
            tagged = kkma.pos(textfile)
            result = '\n'.join([f"{word}_{pos}" for word, pos in tagged])

            # 분석 결과 파일명 설정 (예: original.txt -> original_tagged.txt)
            output_filename = filename.replace('.txt', '_tagged.txt')
            output_path = os.path.join(output_dir, output_filename)

            # 개별 결과 파일 저장
            with open(output_path, 'w', encoding='utf-8') as res_file:
                res_file.write(result)
        except Exception as e:
            print(f"오류 발생: {filename} 처리 중 - {e}")
            continue # 다음 파일로 넘어감

# 3. 결과 파일들을 다시 ZIP으로 압축
with zipfile.ZipFile(final_zip_name, 'w') as res_zip:
    for root, dirs, files in os.walk(output_dir):
        for file in files:
            res_zip.write(os.path.join(root, file), arcname=file)

print(f"\n모든 작업이 완료되었습니다!")
print(f"최종 결과 파일: {final_zip_name}")
```
